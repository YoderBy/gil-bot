```python
import json
import litellm
from dotenv import load_dotenv
import os
load_dotenv()
def get_vertex_credentials():
    """
    Get the Vertex AI credentials from the environment variable.
    """
    print("Getting Vertex AI credentials at", os.environ.get('GOOGLE_APPLICATION_CREDENTIALS'))
    vertex_credentials_path = os.environ.get('GOOGLE_APPLICATION_CREDENTIALS')
    with open(vertex_credentials_path, 'r') as file:
        vertex_credentials = json.load(file)
    vertex_credentials_json = json.dumps(vertex_credentials)
    return vertex_credentials_json


import litellm
from dotenv import load_dotenv
import os
load_dotenv()


llm = litellm.completion(
    model="gemini-1.5-pro",  # Use the vertex_ai/ prefix
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "What is the capital of the moon?"}
    ]
)

print(llm.choices[0].message.content)
```
